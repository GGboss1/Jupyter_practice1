# Group by aggregation

```python
df = pd.read_csv("C:/Users/ggabi/MyStuff/python_practice/Panda_tutorial/Flavors.csv")
df
```

```python

#group by the base flavor
group_by_frame = df.groupby('Base Flavor') 


#gets the mean of the 
group_by_frame.mean(numeric_only=True)

#will count the number of rows
group_by_frame.count()

#gets the min value of base flavor for each column (alphabetical order)
group_by_frame.min()

#gets the max value of base flavor for each column (reverse alphabetical order)
group_by_frame.max()

#gets the sum/total value of base flavor for each column order)
group_by_frame.sum(numeric_only=True)


#Aggregate (doing multiple of above for a column)
group_by_frame.agg({'Flavor Rating' : ['mean' , 'max' , 'count' , 'sum'], 'Texture Rating':['mean' , 'max' , 'count' , 'sum']})

```

### Multiple groups

```python
#group by base flavor and liked
df.groupby(['Base Flavor', 'Liked']).mean(numeric_only=True)

df.groupby(['Base Flavor', 'Liked']).mean(numeric_only=True).agg({'Flavor Rating' : ['mean' , 'max' , 'count' , 'sum'], 'Texture Rating':['mean' , 'max' , 'count' , 'sum']})

```

```python
#shortcut for all(less specific)
group_by_frame.describe()
```

# Merging dataframes

```python
df1 = pd.read_csv("C:/Users/ggabi/MyStuff/python_practice/Panda_tutorial/LOTR.csv")

df1
```

```python
df2 = pd.read_csv("C:/Users/ggabi/MyStuff/python_practice/Panda_tutorial/LOTR 2.csv")

df2
```

### Joins

##### merge inner join

```python
#df1 is left dataframe #df2 is right
#merge is good for joining based on column names
#default is inner join (used automatic values for join)
#(reminder : inner join only return a join of matching values)
df1.merge(df2)

#inner join
df1.merge(df2, how = 'inner')

#will merge using FellowshipID(there will be 2 firstname values for each table since not merging based on firstname )
df1.merge(df2, how = 'inner', on = 'FellowshipID')

#will merge using FellowshipID and first name(what happened automatically when using df1.merge(df2))
df1.merge(df2, how = 'inner', on = ['FellowshipID','FirstName'])
```

##### merge outer join

```python

#outer join
#reminder : / outer join returns everything --> if no value will leave empty or null or NaN(not a number))
df1.merge(df2, how = 'outer')

# will try to bring data from table right table to left table (left table is base table)
df1.merge(df2, how = 'left')

# will try to bring data from table left table to right table (right table is base table)
df1.merge(df2, how = 'right')


#will join by comparing each value from both tables to each other(not really used)
#df1.merge(df2, how = 'cross')

```

#### join(.join)

```python
#merge is good for joining based on index
df1.join(df2, on= 'FellowshipID', how = 'outer', rsuffix= '_Right')


#extra steps needed for join:
#created a new index for both data sets to join to
#stated which one is left and which one is right
df4 = df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'), lsuffix= '__Left', rsuffix= '_Right', how = 'outer')
df4
```
